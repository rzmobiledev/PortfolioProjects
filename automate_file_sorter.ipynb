# **File Sorter**

```python
import os
import shutil
```

```python
path = r"C:/Users/MSi/Downloads/"
file_names = os.listdir(path)
```

```python
file_names
```

```python

def create_folder(folderName: str):
    os.makedirs(path + folderName)
    print(f"{folderName} created")

def is_folder_exists(folderName: str) -> bool:
    if(os.path.exists(path + folderName)):
        return True
    return False

def remove_folder(folderName: str):
    folder_path = path + folderName
    if(os.path.exists(folder_path)):
        shutil.rmtree(folder_path)
        print(f"{folderName} removed")
    else:
        print(f"{folderName} is not exist")


def move_all_files_to(folderName: str, file_ext: str):
    if not is_folder_exists(folderName):
        create_folder(folderName)
        
    for pdf in file_names:
        if file_ext in pdf:
            shutil.move(path + pdf, path + folderName + "/" + pdf)
```

```python
move_all_files_to('ALL_PDF', '.pdf')
```

```python
for csv in file_names:
    if '.txt' in csv:
        os.remove(path + csv)
```

```python
move_all_files_to('ALL_IMAGES', '.jpg')
```
